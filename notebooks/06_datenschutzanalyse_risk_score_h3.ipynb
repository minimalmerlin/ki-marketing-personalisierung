## Rechtlicher und ethischer Rahmen

Die Analyse von KI-gestützter Personalisierung im Marketing erfolgt im Spannungsfeld zwischen wirtschaftlichem Nutzen und datenschutzrechtlichen Anforderungen. Insbesondere die Datenschutz-Grundverordnung (DSGVO) setzt klare Grenzen für die Verarbeitung personenbezogener Daten und automatisierte Entscheidungsfindung.

Im Mittelpunkt dieser Arbeit stehen insbesondere Art. 22 DSGVO sowie die Leitlinien des European Data Protection Board (EDPB) und die Guidance des Information Commissioner's Office (ICO).

### Art. 22 DSGVO – Automatisierte Entscheidungen und Profiling

Art. 22 DSGVO regelt den Umgang mit automatisierten Entscheidungen, die rechtliche Wirkung entfalten oder betroffene Personen erheblich beeinträchtigen können. Betroffene haben grundsätzlich das Recht, nicht ausschließlich einer solchen automatisierten Entscheidung unterworfen zu werden.

Im Kontext von Marketing bedeutet dies, dass insbesondere stark personalisierte und automatisierte Systeme kritisch zu bewerten sind, wenn sie auf individuellen Profilen basieren und Entscheidungen ohne menschliche Beteiligung treffen.

Segmentbasierte Ansätze fallen in der Regel nicht unter diese strengen Anforderungen, da sie keine individuellen Entscheidungen auf Personenebene treffen. Individuelle Vorhersagemodelle hingegen können in den Anwendungsbereich von Art. 22 fallen, insbesondere wenn sie zur automatisierten Steuerung von Angeboten oder Preisen eingesetzt werden.

### EDPB-Leitlinien zum Profiling

Die Leitlinien des European Data Protection Board (EDPB) konkretisieren die Anwendung der DSGVO im Kontext von Profiling und automatisierter Entscheidungsfindung.

Profiling wird dabei als jede Form der automatisierten Verarbeitung personenbezogener Daten definiert, die zur Bewertung persönlicher Aspekte einer Person dient, beispielsweise zur Analyse von Verhalten, Interessen oder Präferenzen.

Die Leitlinien unterscheiden implizit zwischen verschiedenen Intensitätsstufen des Profilings. Während aggregierte oder segmentbasierte Analysen als weniger kritisch gelten, steigt das Risiko bei individueller, detaillierter Profilbildung deutlich an.

### ICO Guidance zu Direct Marketing und KI

Die Guidance des Information Commissioner's Office (ICO) ergänzt die DSGVO-Perspektive durch praxisnahe Empfehlungen für Unternehmen.

Zentrale Prinzipien sind:

- Transparenz gegenüber Nutzern über die Verwendung ihrer Daten
- Zweckbindung und Datensparsamkeit
- Fairness und Vermeidung unerwarteter oder aufdringlicher Personalisierung
- Rechenschaftspflicht bei automatisierten Systemen

Für Marketinganwendungen bedeutet dies, dass einfache Personalisierungsmaßnahmen in der Regel unkritisch sind, während hochgradig individualisierte und automatisierte Entscheidungslogiken eine erhöhte regulatorische Aufmerksamkeit erfordern.

### Einordnung für das Projekt

Die im Projekt untersuchten Personalisierungsansätze lassen sich entlang dieser regulatorischen Perspektiven einordnen.

- Generische und segmentbasierte Ansätze bewegen sich überwiegend im Bereich geringer bis moderater Risiken.
- Segmentbasierte Empfehlungssysteme stellen eine Zwischenstufe dar, da sie bereits auf Nutzerverhalten basieren.
- Individuelle Vorhersagemodelle erreichen hingegen eine Stufe, die dem im Sinne der DSGVO relevanten Profiling nahekommt.

Diese Einordnung bildet die Grundlage für die spätere Entwicklung des Risk Scores sowie die Bewertung der Hypothese H3.

## Bewertungslogik: Risk Score

Zur systematischen Bewertung der Datenschutzrisiken wurde ein Risk Score entwickelt. Der Score basiert auf drei Dimensionen:

- **Data Volume**: Umfang und Vielfalt der genutzten Daten
- **Sensitivity**: Sensitivität der verwendeten Merkmale
- **Granularity**: Grad der Individualisierung der Ansprache

Formel:

Risk Score = 0.3 × Data Volume + 0.3 × Sensitivity + 0.4 × Granularity

Die resultierenden Werte liegen zwischen 0 und 1:
- 0.0–0.3 = niedrig
- 0.3–0.7 = mittel
- 0.7–1.0 = hoch

## Profiling-Level

Zur Einordnung der Personalisierungsintensität werden drei Stufen unterschieden:

- **Low**: generische Ansprache ohne individuelle Auswertung
- **Medium**: segmentbasierte Ansprache auf Gruppenebene
- **High**: individuelle, prädiktive Modelle auf Personenebene

## Trade-off zwischen Performance und Datenschutzrisiko

Die folgende Darstellung zeigt, wie sich mit steigender Personalisierung sowohl die Performance als auch das Datenschutzrisiko verändern.

```python
import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------
# 1) Daten (FINAL)
# -----------------------------
data = pd.DataFrame({
    "Methode": [
        "Generisch",
        "Mass Market",
        "Experienced",
        "Passive",
        "Champions",
        "Individuelles Modell"
    ],
    "Risk_Score": [
        0.20,
        0.60,
        0.60,
        0.60,
        0.60,
        0.90
    ],
    "Performance": [
        11.7,
        11.0,
        26.0,
        0.1,
        10.6,
        94.31   # AUC umgerechnet
    ],
    "Datensatz": [
        "Bank",
        "Bank",
        "Bank",
        "GA4",
        "GA4",
        "Model"
    ]
})

# -----------------------------
# 2) Plot
# -----------------------------
plt.figure(figsize=(10, 6))

for dataset in data["Datensatz"].unique():
    subset = data[data["Datensatz"] == dataset]
    plt.scatter(
        subset["Risk_Score"],
        subset["Performance"],
        label=dataset,
        s=120
    )

# -----------------------------
# 3) Labels anpassen
# -----------------------------
for _, row in data.iterrows():
    
    if row["Methode"] == "Champions":
        plt.annotate(
            row["Methode"],
            (row["Risk_Score"], row["Performance"]),
            xytext=(6, -12),
            textcoords="offset points"
        )
        
    elif row["Methode"] == "Individuelles Modell":
        plt.annotate(
            row["Methode"],
            (row["Risk_Score"], row["Performance"]),
            xytext=(6, 6),
            textcoords="offset points"
        )
        
    else:
        plt.annotate(
            row["Methode"],
            (row["Risk_Score"], row["Performance"]),
            xytext=(6, 6),
            textcoords="offset points"
        )

# -----------------------------
# 4) Layout
# -----------------------------
plt.xlabel("Risk Score")
plt.ylabel("Performance (%)")
plt.title("Trade-off zwischen Datenschutzrisiko und Performance")

plt.legend()
plt.grid(True)
plt.xlim(0, 1.0)
plt.ylim(0, max(data["Performance"]) + 10)

# -----------------------------
# 5) Speichern
# -----------------------------
plt.savefig("tradeoff_final_complete.png", bbox_inches="tight", dpi=300)

plt.show()
```

### Abbildung: Trade-off zwischen Performance und Datenschutzrisiko

Die Abbildung zeigt den Zusammenhang zwischen dem Grad der Personalisierung (gemessen über den Risk Score) und der Performance der jeweiligen Marketingansätze.

Auf der x-Achse ist der Risk Score dargestellt, der das datenschutzrechtliche Risiko der jeweiligen Methode abbildet. Auf der y-Achse ist die Performance dargestellt, gemessen anhand unterschiedlicher Metriken wie Response Rate, Conversion Rate und Precision@5.

Die Ergebnisse verdeutlichen, dass mit zunehmender Personalisierung sowohl die Performance als auch das Datenschutzrisiko ansteigen. Während generische Ansätze ein geringes Risiko aufweisen, erzielen sie nur moderate Ergebnisse. Segmentbasierte Ansätze zeigen bereits deutliche Verbesserungen, insbesondere im Segment „Experienced“.

Der segmentbasierte Recommender erhöht die Relevanz der Empfehlungen erheblich, was sich in einer deutlich höheren Precision@5 widerspiegelt. Die höchste Performance wird durch das individuelle Response-Modell erreicht, das eine AUC von 0.9431 aufweist.

Zur besseren Vergleichbarkeit wurde die AUC des Modells in Prozent umgerechnet (94.31) und gemeinsam mit den anderen Performance-Metriken dargestellt. Es ist jedoch zu beachten, dass es sich hierbei um unterschiedliche Kennzahlen handelt, die nur eingeschränkt direkt vergleichbar sind.

Insgesamt macht die Abbildung den zentralen Trade-off des Projekts sichtbar: Höhere Personalisierung führt zu besseren Ergebnissen, geht jedoch mit einem deutlich erhöhten Datenschutzrisiko einher. Besonders das individuelle Modell markiert dabei den Übergang in einen kritischeren Bereich im Sinne der DSGVO.

# Interpretation und Diskussion der Ergebnisse

In diesem Abschnitt werden die Ergebnisse der Datenschutzanalyse interpretiert und in einen größeren Kontext eingeordnet. Ziel ist es, die Implikationen der beobachteten Zusammenhänge zwischen Personalisierungsgrad, Performance und Datenschutzrisiko zu verstehen.

## Zusammenhang zwischen Personalisierung und Performance

Die Ergebnisse zeigen, dass mit zunehmendem Grad der Personalisierung die Performance der Marketingmaßnahmen deutlich steigt. Während generische Ansätze nur begrenzte Wirkung erzielen, führen segmentbasierte Methoden bereits zu erheblichen Verbesserungen.

## Zusammenhang zwischen Personalisierung und Datenschutzrisiko

Parallel zum Anstieg der Performance zeigt sich ein kontinuierlicher Anstieg des Datenschutzrisikos. Dieser ergibt sich aus drei zentralen Faktoren: dem Umfang der genutzten Daten, der Sensitivität der Merkmale und dem Grad der Individualisierung.

Während generische Ansätze nur ein sehr geringes Risiko aufweisen, steigt dieses bei segmentbasierten Methoden moderat an. Besonders deutlich wird der Anstieg jedoch bei individuellen Vorhersagemodellen, die eine Vielzahl personenbezogener und verhaltensbezogener Daten kombinieren.

Das verwendete Logistic-Regression-Modell stellt hierbei die höchste Risikostufe dar. Mit einem Risk Score von 0.90 basiert es auf detaillierten Nutzerprofilen und ermöglicht eine individuelle Vorhersage von Verhaltenswahrscheinlichkeiten. Dadurch entsteht ein hohes Maß an Profiling, das datenschutzrechtlich relevant ist.

## Trade-off zwischen Nutzen und Risiko

Die zentrale Erkenntnis des Projekts ist das Bestehen eines Trade-offs zwischen Performance und Datenschutzrisiko. Mit zunehmender Personalisierung steigen sowohl der Nutzen für das Unternehmen als auch die potenziellen Risiken für die Privatsphäre der Nutzer.

Während segmentbasierte Ansätze einen guten Kompromiss darstellen, da sie deutliche Performance-Gewinne bei moderatem Risiko ermöglichen, verschiebt sich dieses Gleichgewicht bei individuellen Modellen. Hier wird zwar die höchste Performance erreicht, gleichzeitig steigt jedoch auch das Risiko auf ein Niveau, das zusätzliche regulatorische Anforderungen erforderlich macht.

Die Trade-off-Analyse zeigt somit, dass der Mehrwert von Personalisierung nicht isoliert betrachtet werden kann, sondern immer im Kontext der damit verbundenen Risiken bewertet werden muss.

## Beantwortung der Hypothese H3

Die Hypothese H3 postuliert, dass ab einem bestimmten Grad der Individualisierung die datenschutzrechtlichen und ethischen Risiken den Mehrwert überwiegen.

Die Ergebnisse des Projekts legen nahe, dass diese kritische Schwelle nicht bereits bei segmentbasierter Personalisierung erreicht wird. Vielmehr zeigt sich, dass segmentbasierte Ansätze einen signifikanten Mehrwert bieten, ohne ein hohes Risiko zu erzeugen.

Erst bei individuellen, prädiktiven Modellen wird diese Grenze relevant. Diese Modelle nutzen umfangreiche personenbezogene Daten und erstellen detaillierte Profile einzelner Nutzer. Dadurch steigt das Risiko auf ein Niveau, bei dem regulatorische Anforderungen, insbesondere im Kontext von Art. 22 DSGVO, berücksichtigt werden müssen.

H3 kann somit als teilweise bis weitgehend bestätigt angesehen werden: Die kritische Grenze liegt nicht im Bereich einfacher Personalisierung, sondern bei tiefgehenden, individualisierten Profiling-Ansätzen.

## Handlungsempfehlungen

Aus den Ergebnissen lassen sich mehrere konkrete Empfehlungen für Unternehmen ableiten:

**1. Segmentbasierte Personalisierung gezielt einsetzen**  
Segmentbasierte Ansätze bieten einen hohen Mehrwert bei vergleichsweise geringem Risiko. Sie stellen daher in vielen Fällen den optimalen Einstieg in datengetriebenes Marketing dar.

**2. Individuelle Modelle nur unter klaren Rahmenbedingungen nutzen**  
Individuelle Vorhersagemodelle sollten nur dann eingesetzt werden, wenn eine klare rechtliche Grundlage, transparente Kommunikation und geeignete Governance-Strukturen vorhanden sind.

**3. Datenschutz frühzeitig integrieren (Privacy by Design)**  
Datenschutz sollte nicht erst nachträglich berücksichtigt werden, sondern bereits bei der Entwicklung von Modellen und Kampagnen eine zentrale Rolle spielen.

**4. Transparenz gegenüber Nutzern erhöhen**  
Nutzer sollten nachvollziehen können, warum sie bestimmte Inhalte oder Angebote erhalten. Dies stärkt Vertrauen und reduziert regulatorische Risiken.

**5. Datenumfang bewusst begrenzen (Datensparsamkeit)**  
Nicht jede verfügbare Information sollte genutzt werden. Eine gezielte Auswahl relevanter Merkmale kann helfen, das Risiko zu reduzieren, ohne die Performance wesentlich zu beeinträchtigen.

**6. Trade-off aktiv managen**  
Unternehmen sollten Personalisierung nicht maximal, sondern optimal einsetzen. Ziel ist es, ein Gleichgewicht zwischen Nutzen und Risiko zu finden.
